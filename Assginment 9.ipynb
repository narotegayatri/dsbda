```python
import seaborn as sns
```

```python
data = sns.load_dataset('titanic')
data
```

```python
sns.boxplot(data = data, x='sex', y='age', hue='survived')
```

```python
sns.boxplot(data=data,x='sex',y='age')
```

```python
data_survived = data.where(data['survived']==1).dropna()
data_survived.head(5)
```

```python
sns.boxplot(data=data_survived,x='sex',y='age')
```

```python
data_not_survived = data.where(data['survived']!=1).dropna()
data_not_survived.head(5)
```

```python
sns.boxplot(data=data_not_survived,x='sex',y='age')
```
