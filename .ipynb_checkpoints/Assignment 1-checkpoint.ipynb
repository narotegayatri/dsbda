```python
import pandas as pd
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

## Description of Columns/Features

- gender: Gender of the student (M/F)
- race/ethnicity: Student's race or ethnicity (Group A - E)
- parental level of education: Education level of parents
- lunch: Lunch service used by student
- test preparation course: If student has completed test preparation
- math score: Marks scored by student in Maths
- reading score: Marks scored by student in reading
- writing score: Marks scored by student in writing
